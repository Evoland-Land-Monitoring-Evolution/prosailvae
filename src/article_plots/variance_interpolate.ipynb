```python
import numpy as np
import matplotlib.pyplot as plt

def std_interpolate(x0, std0, x1, std1, x):
    assert (x <= x1).all() and (x >= x0).all()
    u0 = (x1-x)/(x1-x0)
    u1 = (x-x0)/(x1-x0)
    return np.sqrt((u0 * std0)**2 + (u1*std1)**2)

def interpolate(x0, std0, x1, std1, x):
    assert (x <= x1).all() and (x >= x0).all()
    u0 = (x1-x)/(x1-x0)
    u1 = (x-x0)/(x1-x0)
    return u0 * std0 + u1 * std1

dx = 0.05
x0 = 0
x1 = 1
std0 = 1
std1 = 1

x = np.arange(x0, x1, dx)
inter_std = std_interpolate(x0, std0, x1, std1, x)
inter = interpolate(x0, std0, x1, std1, x)
plt.figure()
plt.scatter(x, inter_std)
plt.scatter([x0, x1], [std0, std1])
plt.figure()
plt.scatter(x, inter)
plt.scatter([x0, x1], [std0, std1])
```

```python
def average_variance(var1, var2):
    return (var1 + var2) / 4
print(np.sqrt(average_variance(1, 1)))
```

```python
n=500
y1 = np.random.rand(n)
y0 = np.random.rand(n)
x = np.arange(x0, x1, dx)
plt.figure()
for i in range(n):
    plt.plot(x, interpolate(x0, y0[i], x1, y1[i], x), lw=0.5)
```
